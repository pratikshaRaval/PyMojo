```python
import pandas as pd
df = pd.read_csv('books.csv', index_col='title')
df.head()
```

# Finding 1: Top 10 books with the highest average_rating and more than 50 written text reviews

```python
#top 10 books with the highest average_rating where there are more than 50 written text reviews
df[df.text_reviews_count>50][['average_rating','text_reviews_count']].sort_values(by=['average_rating','text_reviews_count'], ascending=[0,0]).head(10)
```

# Finding 2: Top 10 books with the highest text_reviews_count

```python
#top 10 books with highest text_reviews_count
most_reviews = df[['authors', 'text_reviews_count']].sort_values('text_reviews_count', ascending = False).head(10)
most_reviews
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.barplot(most_reviews['text_reviews_count'], most_reviews.index, palette='rocket')
```
